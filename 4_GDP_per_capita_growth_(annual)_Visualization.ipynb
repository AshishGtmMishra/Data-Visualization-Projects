```python
import pandas as pd


# Loading the data
filepath = '/content/sample_data/GDP_Per_capita_Growth_anually_.csv'
df = pd.read_csv(filepath)

df.head(10)
```

```python
import matplotlib.pyplot as plt

# Replacing ".." with NaN
# Convert the year columns to numeric, errors='coerce'
for col in df.columns[4:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop the non-numeric columns
numeric_df = df.drop(columns=['Series Name', 'Series Code', 'Country Name', 'Country Code'])

# Plotting the distribution of GDP per capita growth for all years
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_df.columns):
    plt.subplot(3, 4, i+1)
    numeric_df[column].hist(bins=20)
    plt.title('Distribution in ' + column)
    plt.xlabel('GDP per capita growth (%)')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The plots above show the distribution of GDP per capita growth rates for the years where data is available. Each subplot represents a different year, with the x-axis showing the growth rate percentage and the y-axis indicating the frequency of countries within each growth rate bin. The histograms provide a quick way to assess the spread and central tendency of GDP growth rates across the sampled countries for each year.

```python
import seaborn as sns

# Melting the dataframe to have years and values in the seperate columns
melted_df = df.melt(id_vars=['Country Name', 'Country Code'],
                     var_name='Year', value_name='GDP Growth (%)',
                     value_vars=df.columns[4:])

# Removing rows with NaN values in 'GDP Growth (%)'
melted_df = melted_df.dropna(subset=['GDP Growth (%)'])
```

```python
# Convert 'Year' to a string
melted_df['Year'] = melted_df['Year'].str.extract('(\d{4})')

# Convert 'Year' to integer
melted_df['Year'] = melted_df['Year'].astype(int)
```

```python
# Plotting GDP Growth over the years for all countries
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_df, x='Year', y='GDP Growth (%)', estimator='mean', ci=None)
plt.title('Average GDP per Capita Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

- By looking at the above visualization we can see that there we a sudden dip in the Average GDP Growth in the year 2020, which indicates that there was a severe impact on economy suring the corona period.

- Looking at the pervious trend on year before 2000 it can be seen that there was a growing trend in the GDP growth, which might have happend because of robust urbanization in the Asian countries like China, Japan, South Korea etc.

- But the dropdown indicates that there was a decrease in the GDP percentage after the year 2000, it may have happend because of the recession that was caused in Japan and some other asian countries, which was in a down fall till 2012-2013, and this might explain the impact of the Great Recession of 2008 on the Global GDP average.

- After the year 2021 the world has recovered from the impacts of Lockdown and is in the progress of positve growth.
